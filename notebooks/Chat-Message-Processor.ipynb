## Upload and Concatenate CSV Files

```python
import pandas as pd
import glob
import os

# Define the path where your CSV files are stored
file_path = './discord-chat-messages/*.csv'

# Use glob to get a list of all CSV files in the specified path
files = glob.glob(file_path)

# Initialize an empty list to store DataFrames
dfs = []

def is_empty(file_path):
    return os.stat(file_path).st_size == 0

# Loop through the files and read them into DataFrames
for file in files:
    if is_empty(file):
        continue
    df = pd.read_csv(file)
    dfs.append(df)

# Concatenate all DataFrames in the list
concatenated_df = pd.concat(dfs, ignore_index=True)

# Now, `concatenated_df` contains all the data from your CSV files
concatenated_df
```
